# I. Leitura de Dados

Para iniciarmos este projeto, efetuamos a leitura do arquivo *csv* fornecido pelo cliente, criando, assim, nosso **dataframe** (df).

```python
df = pd.read_csv('dados_TREINO.csv')
df
```

# II. Filtro de Dados

Após análise primária da superfície dos dados, iniciamos o processo de filtragem do **dataframe**, separando as informações que não serão utilizadas ou que possam interferir neste estudo, sejam por imprecisão, sejam por poluição.

```python
# df.dropna() é uma função que remove as linhas com valores nulos
# inplace=True é para que a função altere o dataframe original
df.dropna(inplace=True)
df
```

# IV. Seleção de Atributos

**handles.nhandles** - O número de handles abertos por um processo pode ajudar a identificar a presença de processos maliciosos que estão usando muitos recursos do sistema.

**dlllist.ndlls** - A lista de DLLs carregadas por um processo pode ajudar a identificar as bibliotecas utilizadas pelo software malicioso, especialmente se elas forem diferentes das DLLs associadas a software legítimo.

**pslist.avg_threads** - O número médio de threads por processo pode ajudar a identificar se um processo está tentando fazer muitas coisas ao mesmo tempo, o que pode ser uma característica comum de malwares.

**ldrmodules.not_in_load** - O número de módulos não carregados pode ajudar a identificar se um processo está tentando ocultar as bibliotecas que está usando para evitar a detecção.

```python
# df.loc[: [<coluna1>, <coluna2>]]: isso seleciona apenas as colunas desejadas
df_filtrado = df.loc[:, ['dlllist.ndlls',   'pslist.avg_threads',
                         'ldrmodules.not_in_load', 'handles.nhandles',
                         # Essas duas precisam estar na solução final:
                         'Category', 'Class'
                         ]]

df_filtrado
```

```python
# Realizar treino para classificar se é malware ou benigno
# 1. Separar os dados em X e y
# 2. Separar os dados em treino e teste
# 3. Treinar o modelo
# 4. Testar o modelo
# 5. Avaliar o modelo

X = df_filtrado.drop(columns=['Category', 'Class'])
y = df_filtrado['Category']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

print(f'X_train: {X_train.shape}')
print(f'X_test: {X_test.shape}')

print(f'y_train: {y_train.shape}')
print(f'y_test: {y_test.shape}')
```

```python
classes = list(y_train.unique())
classes.sort()

classes # confirmado, problema não é multi-classe (Mais de 2 classes), e sim binário (2 classes)
```

```python
clf = DecisionTreeClassifier().fit(X_train, y_train)
clf
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)

print(f"Acurácia: {round(accuracy, 4) * 100}%")
```

```python
# Utilizando outro método de treino
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier().fit(X_train, y_train)
clf
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)

print(f"Acurácia: {round(accuracy, 4) * 100}%")
```

```python
# Utilizando outro método de treino
from sklearn.svm import SVC

clf = SVC().fit(X_train, y_train)
clf
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)

print(f"Acurácia: {round(accuracy, 4) * 100}%")
```

```python
# Utilizando outro método de treino
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression().fit(X_train, y_train)
clf
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)

print(f"Acurácia: {round(accuracy, 4) * 100}%")
```
